1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
Were the passengers in higher classes more likely to survive?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://www.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
titanic = pd.read_csv(url)

# Visualize the survival rate by ticket class
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic, ci=None)
plt.title('Survival Rate by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Survival Rate')
plt.show()
```

5. Create a grouped bar chart to visualize the survival rates of passengers across different age groups and genders. Were certain age groups or genders more likely to survive?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
titanic = pd.read_csv(url)

# Create age groups
bins = [0, 12, 18, 35, 60, 80]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

# Create a grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='AgeGroup', y='Survived', hue='Sex', data=titanic, ci=None)
plt.title('Survival Rates by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.legend(title='Gender')
plt.show()
```
